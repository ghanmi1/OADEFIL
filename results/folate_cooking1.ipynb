```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


# Define the function to calculate concentration at time t
def concentration(t, C0, k):
    return C0 * np.exp(-k * t)

# Define the time range
t_values = np.linspace(0, 150)  # Adjust the time range as needed

# Define the initial concentration
C0 = 1.0  # Initial concentration, adjust as needed

# Define the rate constants (k values) for different pH and temperatures for 5-Methyl tetrahydrofolic acid
rate_constants_5mthf = {
    3: {60: 0.0003, 80: 0.014, 100: 0.028},
    5: {60: 0.0033, 80: 0.022, 100: 0.044},
    7: {60: 0.0064, 80: 0.014, 100: 0.034},
}

# Define rate constants (k values) for Tetrahydrofolic acid, Folic acid, and 10-Formyl folic acid for different pH and temperatures
# You can adjust these values as needed
thf_rate_constants = {
    3: {60: 0.0046, 80: 0.0081, 100: 0.017},
    5: {60: 0.002, 80: 0.0019, 100: 0.0014},
    7: {60: 0.00003, 80: 0.00004, 100: 0.00005},
}



# Plot concentration versus time for each pH and temperature
for ph, temp_k_values in rate_constants_5mthf.items():
    plt.figure(figsize=(10, 8))
    i = 1  # Subplot index
    for temp, k in temp_k_values.items():
        Ct_5mthf = concentration(t_values, C0, k)
        plt.subplot(3, 3, i)
        plt.plot(t_values, Ct_5mthf, label=f'5THF-{temp}°C')
        
        # Plot additional data for Tetrahydrofolic acid
        thf_k = thf_rate_constants[ph][temp]
        thf_Ct = concentration(t_values, C0, thf_k)
        plt.plot(t_values, thf_Ct, label=f'Tetra-{temp}°C', linestyle='--')
        
        plt.xlabel('Time (t)')
        plt.ylabel('Concentration (C)')
        plt.title(f'C vs. T at pH {ph}')
        plt.legend()
        plt.grid(True)
        i += 1
    plt.tight_layout()
    plt.show()


```
