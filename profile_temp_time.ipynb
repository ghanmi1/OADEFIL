```python
import matplotlib.pyplot as plt
import numpy as np

# Durée de la phase stable en minutes
t_stable = 10  

# Fonction température-temps
def temperature_profile(t, t_stable):
    if t <= 50:
        return 25 + 2 * t
    elif t <= 50 + t_stable:
        return 125
    elif t <= 83 + t_stable:
        return 125 - 3 * (t - 50 - t_stable)
    else:
        return 25

# Générer des points de temps pour tracer le profil
time_points = np.linspace(0, 83 + t_stable, 500)
temperature_points = [temperature_profile(t, t_stable) for t in time_points]

# Tracer le profil température-temps
plt.figure(figsize=(10, 6))
plt.plot(time_points, temperature_points, marker='o')
plt.title('Profil Température-Temps')
plt.xlabel('Temps (minutes)')
plt.ylabel('Température (°C)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initial conditions
initial_temp = 25
final_temp = 125
increase_rate = 2  # degrees per minute
decrease_rate = 3  # degrees per minute
stable_duration = 10  # duration of the stable phase in minutes

# Calculate the duration of the heating phase
heating_duration = (final_temp - initial_temp) // increase_rate

# Calculate the duration of the cooling phase
cooling_duration = (final_temp - initial_temp) // decrease_rate

# Create the time array
time_heating = list(range(heating_duration + 1))
time_stable = list(range(time_heating[-1] + 1, time_heating[-1] + 1 + stable_duration))
time_cooling = list(range(time_stable[-1] + 1, time_stable[-1] + 1 + cooling_duration + 1))

# Create the temperature array for each phase
temp_heating = [initial_temp + increase_rate * t for t in time_heating]
temp_stable = [final_temp] * len(time_stable)
temp_cooling = [final_temp - decrease_rate * (t - time_stable[-1]) for t in time_cooling]

# Combine time and temperature arrays
time = time_heating + time_stable + time_cooling
temperature = temp_heating + temp_stable + temp_cooling

# Plot the temperature-time profile
plt.figure(figsize=(10, 6))
plt.plot(time, temperature, marker='o')
plt.title('Temperature-Time Profile')
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```
